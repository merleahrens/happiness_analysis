**Teil 1. Einflüsse auf Lebenszufriedenheit innerhalb eines Landes (WITHIN)**

**Autokorrelation**  
Da es sich um Paneldaten (Länder × Jahre) handelt, ist Autokorrelation der Residuen zu erwarten (Länder haben ähnliche Glückswerte über Jahre).   

Ich schätze daher Modelle mit Länder-Fixed-Effects, kontrolliere für Jahreseffekte und verwende cluster-robuste Standardfehler auf Länderebene. 
Dadurch wird die Unabhängigkeit der Fehler realistisch angenähert und die Inferenz (Standardfehler, p‑Werte) robuster.  
    
**Fixed Effects pro Land**  
Das konstante Glücks-Grundlevel pro Land als Mittelwert-Linie abziehen und somit nur noch die Veränderung innerhalb eines Landes über Zeit betrachten.  
**Fixed Effects für Zeit**  
Effekte, die alle Länder im selben Jahr betreffen (z.B. Covid, globale Finanzkrisen).  
**Cluster-robuste Standardfehler**  
Erlauben, dass Fehler innerhalb eines Clusters (Land) korreliert sind, aber zwischen Clustern (verschiedenen Ländern) unabhängig.  

**A. Regressionsanalyse BIP - Lebenszufriedenheit**

**Gesamtes BIP-Modell mit Cluster**  
Annahme: Fehler INNERHALB Land korrelieren, ZWISCHEN Ländern nicht.  
Länder-Cluster erzeugen.  

Steigt das Glück in einem Land im Jahr t relativ zum weltweiten Jahresdurchschnitt, 
wenn dort das BIP relativ zum weltweiten Jahresdurchschnitt steigt?

```python
# Fixed Effects: Länder-Dummies & Zeit-Dummies (Land und Jahr als kategorische C-Variable) 
# und Clustering (Cluster-robuste Standardfehler)

# Vorbereitung: Länder als integer codieren
land_codes = df['Land'].factorize()[0]

# Modell, Results und Summary
model_BIP_FE_time = smf.ols('Lebenszufriedenheit ~ BIP_pro_Kopf + C(Land) + C(Jahr)', data=df)
results_cluster = model_BIP_FE_time.fit(cov_type='cluster', cov_kwds={'groups': land_codes})
results_cluster.summary()
```

**Ergebnisse: Wie hängen BIP und Lebenszufriedenheit zusammen?**  
Das Fixed-Effects-Panelmodell mit cluster-robusten Standardfehlern erklärt 90,9% der Varianz in der Lebenszufriedenheit über 1882 Beobachtungen (2011-2024) hinweg.  

Der BIP_pro_Kopf-Koeffizient zeigt einen positiven Trend-Effekt (β = 0.000019, p = 0.097): Innerhalb jedes Landes steigt die Lebenszufriedenheit tendenziell bei höherem Pro-Kopf-BIP (10.000 USD → +0.19 Glückspunkte), ist jedoch nicht ganz signifikant (p > 0.05).

Die cluster-robusten SE adressieren Länder-spezifische Autokorrelation erfolgreich, was die Robustheit der Schätzung unterstreicht.

**B. Regressionsanalyse Demokratie_Score - Lebenszufriedenheit**

**Gesamtes Demokratie_Score-Modell mit Cluster**  
Annahme: Fehler INNERHALB Land korrelieren, ZWISCHEN Ländern nicht.  
Länder-Cluster erzeugen.  

Was passiert mit Lebenszufriedenheit, wenn sich DIESELBE DEMOKRATIE in einem Land über Jahre ändert?

```python
# Modell, Results und Summary
model_Demokratie_FE_time = smf.ols('Lebenszufriedenheit ~ Demokratie_Score + C(Land) + C(Jahr)', data=df)
results_cluster2 = model_Demokratie_FE_time.fit(cov_type='cluster', cov_kwds={'groups': land_codes})
results_cluster2.summary()
```

**Ergebnisse:**  

Das Fixed-Effects-Modell (cluster-robuste SE) erklärt 91,5% der Varianz (R²=0.915, N=1882, 2011-2024).

Within-Effekt (innerhalb Länder): Demokratie_Score nicht signifikant (β = -0.0030, p = 0.443). Höhere Demokratie-Qualität korreliert nicht mit gesteigerter Lebenszufriedenheit innerhalb einzelner Länder über Zeit.

Erklärung: Demokratien ändern sich langsam → wenig Within-Varianz → keine Signifikanz. Länder-Fixed-Effects fangen die zwischenländlichen Unterschiede ab (90%+ der Varianz).

**C. Regressionsanalyse BIP & Demokratie_Score - Lebenszufriedenheit**

**Multivariates Regressionsmodell für WITHIN Effekte**  
Die Korrelation zwischen Demokratie-Score und Frauenrechte-Index (r=0.84)
deutet auf konzeptionelle Überlappung hin. Aus theoretischen Gründen
(Demokratie als grundlegender Freiheits‑Indikator) wird neben BIP_pro_Kopf nur Demokratie_Score
ins Modell aufgenommen.  

```python
multi = smf.ols('Lebenszufriedenheit ~ BIP_pro_Kopf + Demokratie_Score + C(Land) + C(Jahr)', data=df)
results_multi = multi.fit(cov_type='cluster', cov_kwds={'groups': land_codes})
results_multi.summary()
```

**WITHIN-Effekte (innerhalb Länder):**  

BIP_pro_Kopf: p=0.090 → Trend positiv, nicht signifikant (p>0.05)  
Demokratie_Score: p=0.323 → nicht signifikant  
Adj. R²=0.909 → 91% der Varianz durch Länder-Fixed-Effects + Cluster-SE erklärt  

| Variable         | p-Wert | Status |
| ---------------- | ------ | ------ |
| BIP_pro_Kopf     | 0.090  | Trend  |
| Demokratie_Score | 0.323  | n.s.   |  

Fazit: Länder-FE erklären 91% – Within-Effekte der Variablen nicht signifikant.  
**Robustheitscheck:** Frauenrechte_Index (p=0.18) wurde in ein Zwischenmodell mit aufgenommen und ergänzte die Hauptergebnisse 
nicht signifikant.**

**Teil 2. Einflüsse auf Lebenszufriedenheit zwischen Ländern (BETWEEN)**

```python
# Erzeugen eines Dataframe mit Länder-Mittelwerten
leben = df.groupby('Land')['Lebenszufriedenheit'].mean()
bip = df.groupby('Land')['BIP_pro_Kopf'].mean()
demo = df.groupby('Land')['Demokratie_Score'].mean()
frau = df.groupby('Land')['Frauenrechte_Index'].mean()

df_land = pd.DataFrame({'Lebenszufriedenheit': leben, 'BIP_pro_Kopf': bip, 'Demokratie_Score': demo, 'Frauenrechte_Index': frau}).reset_index()
df_land.head()
```

```python
df_land.shape
```

```python
# Univariates Regressionsmodell für BIP
model_BIP2 = smf.ols(formula='Lebenszufriedenheit ~ BIP_pro_Kopf', data=df_land)
results = model_BIP2.fit()
results.summary()
```

**BETWEEN Ländervergleich - BIP**  
-BIP erklärt 57% der Varianz (adj. R² = 0.570), das Ergebnis ist hoch signifikant (p = 0.000).  
-keine Hinweise auf Autokorrelation (Durbin-Watson = 2.086).  

Ein Anstieg des BIP pro Kopf um 10.000 USD geht mit 
+0.32 Punkten höherer Lebenszufriedenheit einher 
(β = 0.0000322, p < 0.001).

```python
# Univariates Modell für Demokratie
model_Demokratie2 = smf.ols(formula='Lebenszufriedenheit ~ Demokratie_Score', data=df_land)
results = model_Demokratie2.fit()
results.summary()
```

**BETWEEN Ländervergleich - Demokratie**  
-BIP erklärt 32,6% der Varianz (adj. R² = 0.326), das Ergebnis ist hoch signifikant (p = 0.000).  
-keine Hinweise auf Autokorrelation (Durbin-Watson = 2.139).  

Ein Demokratie-Punkt mehr korrespondiert mit +0.022 Punkten 
Lebenszufriedenheit. Voll-Demokratien sind 0.22 Punkte glücklicher 
als Autokratien (β = 0.0215, p < 0.001).

```python
# Multivariates Regressionsmodell mit BIP und Demokratie_Score (ohne Frauenrechte_Index, da Multikollinearität mit Demokratie_Score)
# Modell für globale Unterschiede, basierend auf Länder-Mittelwerten (ohne zeitliche Komponente)
model_BIP_Demokratie = smf.ols(formula='Lebenszufriedenheit ~ BIP_pro_Kopf + Demokratie_Score', data=df_land)
results = model_BIP_Demokratie.fit()
results.summary()
```

**Ergebnisse BETWEEN Ländervergleich - BIP & Demokratie**  
BIP und Demokratie erklären 63% der Varianz zwischen Ländern (adj. R² = 0.630, p = 0.000). Keine Autokorrelation (Durbin-Watson = 2.094).  

Beide Effekte bleiben signifikant nach gegenseitiger Kontrolle:  

Pro 10.000 USD BIP → +0.27 Glückspunkte (β=0.000027, p<0.001)  
Pro Demokratie-Punkt → +0.011 Glückspunkte (β=0.011, p<0.001)  

Demokratie-Effekt halbiert sich nach BIP-Kontrolle → läuft teilweise über wirtschaftliche Entwicklung.  

Hohe Konditionszahl (Cond. No. = 7.84e+04) durch Skalenunterschiede (BIP: 500-60k USD, Demokratie: 0-100), p-Werte robust. Korrelation BIP-Demokratie (r=0.54) könnte Multikollinearität erklären.  

→ Wirtschaft + Demokratie treiben zwischenländliche Glücksunterschiede.

```python
# Test: Gibt es einen Zusatzeffekt der Frauenrechte?**
# Multivariates Regressionsmodell mit BIP und Demokratie_Score
# Modell für globale Unterschiede, basierend auf Länder-Mittelwerten (ohne zeitliche Komponente)
model_BIP_Demokratie_Frauen = smf.ols(formula='Lebenszufriedenheit ~ BIP_pro_Kopf + Demokratie_Score + Frauenrechte_Index', data=df_land)
results = model_BIP_Demokratie_Frauen.fit()
results.summary()
```

**Fazit Zusatzfaktor Frauenrechte:**  
**Robustheitscheck BETWEEN**  
BIP (+0.27***) und Demokratie (+0.015***) 
bleiben hochsignifikant trotz Frauenrechte-Kontrolle als möglichem Einflussfaktor.    
Frauenrechte selbst nicht signifikant. (p=0.178). Adj. R² steigt minimal auf 63,2%.

**Fazit: Lebenszufriedenheit unter Berücksichtigung von BIP und Demokratie**  
Unterscheidung zwischen within- (FE-Modelle) und between-Effekten (Ländermittelwerte).  

Hauptbefund:  

BETWEEN (zwischen Ländern): BIP (10k USD → +0.27***, p<0.001) und Demokratie (+0.011***, p<0.001) erklären 63% der Varianz (R²=0.63).  
WITHIN (innerhalb Länder): BIP Trend positiv (p=0.090), Demokratie nicht signifikant (p=0.323), R²=0.91 (hauptsächlich Länder-FE).  
In beiden Fällen hat der Frauenrechte_Index keinen signifikanten Einfluss.  

Global: Wirtschaft + Demokratie erklären zwischenländliche Glücksunterschiede.  
Innerhalb Länder: Keine klar signifikanten Effekte – geringe Within-Varianz der Prädiktoren (BIP/Demokratie ändern sich langsam).
